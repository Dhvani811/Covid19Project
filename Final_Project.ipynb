```python
import pandas as pd
```

```python
confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```python
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```python
df1 = pd.read_csv(confirmed_cases_url)
df1
```

```python
df2 = pd.read_csv(deaths_url)
df2
```

```python
df1.columns
```

```python
del df1['Province/State']
del df1['Country/Region']
del df1['Lat']
del df1['Long']

```

```python
del df2['Province/State']
del df2['Country/Region']
del df2['Lat']
del df2['Long']
```

```python
df1
```

```python
confirmed_cases = df1.sum(axis=0)
confirmed_cases.name = 'confirmed_cases'
df3 = pd.DataFrame(confirmed_cases.transpose())
df3
```

```python
df3
```

```python
deaths = df1.sum(axis=0)
deaths.name = 'deaths'
df4 = pd.DataFrame(deaths.transpose())
df4
```

```python
covid_df = pd.concat([df3,df4])
```
