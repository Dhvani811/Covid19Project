```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
# Save the URL as a string
confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```python
# Save the URL as a string

deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```python
# Load the CSV data into a dataframe

df1 = pd.read_csv(confirmed_cases_url)
df1
```

```python
# Load the CSV data into a dataframe

df2 = pd.read_csv(deaths_url)
df2
```

```python
# Check the names of columns
df1.columns
```

```python
# Delete the unneccessary columns 

del df1['Province/State']
del df1['Country/Region']
del df1['Lat']
del df1['Long']

```

```python
# Delete the unneccessary columns 

del df2['Province/State']
del df2['Country/Region']
del df2['Lat']
del df2['Long']
```

```python
df1
```

```python
# Aggregate and sums confirmed cases  on a global level

confirmed_cases = df1.sum(axis=0)
confirmed_cases.index.name = 'dates'
confirmed_cases.name = 'confirmed_cases'

df3 = pd.DataFrame(confirmed_cases)


df3.reset_index()

df3
```

```python
df3.columns
```

```python
# Aggregate and sum deaths on a global level

deaths = df2.sum(axis=0)
deaths.index.name = 'dates'

deaths.name = 'deaths'
df4 = pd.DataFrame(deaths)
df4
```

```python
# Join both the cases and deaths into a single dataframe

covid_df = df3.join(df4)
```

```python
# Save the dates as a column and set it as an index

covid_df["date"]= covid_df.index
covid_df.set_index("date", inplace=True, drop=True)
```

```python
# URL to fetch the daily lows and highs of the YS stock market index "SPY"

response = requests.get(f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=SPY&apikey={API_KEY}&outputsize=full&from=2020-20-22&to=2023-03-09")

data = response.json()

# Get the daily prices from the response
daily_prices = data["Time Series (Daily)"]

us_stocks = pd.DataFrame(columns=['date', 'high', 'low'])

# Loop through each day and get the daily high and low prices
for date, values in daily_prices.items():
    daily_high = values["2. high"]
    daily_low = values["3. low"]

    date_obj = pd.to_datetime(date)
    date_formatted = date_obj.strftime("%-m/%-d/%y")

    new_row = {'date': date_formatted,'high': daily_high, 'low': daily_low}

    # Add it to the us_stocks dataframe
    us_stocks = us_stocks.append(new_row, ignore_index=True)
```

```python
# Rename the columns and set the date column as index
us_stocks = us_stocks.rename(columns={"date": "date", "high": "us_high", "low": "us_low"})
us_stocks.set_index('date', inplace=True, drop=True)
us_stocks
```
